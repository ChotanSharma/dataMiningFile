```python
df = df.drop(["Dependents"], axis=1)

```

```python
df.head()
```

```python
df.columns
```

```python
features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History']
```

```python
df.loc[df["Loan_Status"] == "Y", "Loan_Status"] = 1
df.loc[df["Loan_Status"] == "N", "Loan_Status"] = 0
```

```python
import numpy as np
df['LoanAmount'] = df['LoanAmount'].replace(np.nan, 0)
df.head()
```

```python
y = df['Loan_Status']
X = df[features]
y = y.astype('int')
```

```python
df.head()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import fbeta_score, make_scorer, accuracy_score, precision_score, recall_score
```

```python
# Initialize the base model (Decision Tree)
base_model = DecisionTreeClassifier(random_state=42)
```

```python
bagging_classifier = BaggingClassifier(n_estimators=5,random_state=0)
```

```python
scores = cross_val_score(bagging_classifier, X, y, cv=5, scoring='precision', error_score='raise')
print(scores)
print(scores.mean())
```

```python
scores = cross_val_score(bagging_classifier, X, y, cv=5, scoring='recall', error_score='raise')
print(scores)
print(scores.mean())
```

```python
scores = cross_val_score(bagging_classifier, X, y, cv=5, scoring='accuracy', error_score='raise')
print(scores)
print(scores.mean())
```

```python
f2_scorer = make_scorer(fbeta_score, beta=2)
scores = cross_val_score(bagging_classifier, X, y, cv=5, scoring=f2_scorer, error_score='raise')
print(scores)
print(scores.mean())
```

```python
scores = cross_val_score(bagging_classifier, X, y, cv=5, scoring='recall', error_score='raise')
print(scores)
print(scores.mean())
```
